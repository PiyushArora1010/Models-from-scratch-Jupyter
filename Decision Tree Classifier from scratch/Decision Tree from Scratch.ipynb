**Cost Function**

```python
## Function for entropy
def entropy(l):
    dic = {}
    for i in l:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i]+=1
    sum = 0
    for i in dic:
        p = dic[i] / len(l)
        sum -= p*math.log2(p)
    return sum
```

**Helper Function**

```python
def maximumFreq(array):
  dictionary = {}
  for i in array:
    if i not in dictionary:
      dictionary[i] = 1
    else:
      dictionary[i]+=1
  count = -1
  ans = -1
  for j in dictionary:
    if dictionary[j] > count:
      count = dictionary[j]
      ans = j
  return ans
```

**Decision Tree Class**

```python
class decisionTree:
    def __init__(self, max_depth, classes = 2,th = 0.05, tree_size = 1000):
        self.maxDepth = max_depth
        self.root = []
        if th <= 0.05:
            self.entropyThreshold = 0.05
        else:
            self.entropyThreshold = th
        self.treeNode = [-1,]*tree_size
        self.classes = classes
        self.treeSize = tree_size
        self.thresholdValues = {}
        
    def entropy(self, l):
        dic = {}
        for i in l:
            if i not in dic:
                dic[i] = 1
            else:
                dic[i] += 1
        sum = 0
        for i in dic:
            p = dic[i] / len(l)
            sum -= p*math.log2(p)
        return sum

    def information_gain(self, entropy1, x1, y1, threshold=1):
        split1 = y1[x1 >= threshold].copy()
        split2 = y1[x1 < threshold].copy()
        entr1 = self.entropy(split1)
        entr2 = self.entropy(split2)
        n1 = len(split1)
        n2 = len(split2)
        n = len(y1)
        info_gain = entropy1 - (n1/n)*entr1 - (n2/n)*entr2
        return info_gain

    def best_split(self, x1, y1, index):
        sorted_array = np.sort(x1)
        avg = np.array([sorted_array[i]+sorted_array[i+1]
                       for i in range(len(y1)-1)])
        avg = avg/2
        thresh_best = 0
        inf = -100
        for t in avg:
            temp = self.information_gain(self.entropy(y1), x1, y1, t)
            if temp > inf:
                thresh_best = t
                inf = temp
        print("Index number:", index)
        print("Threshold ==>", thresh_best, "----", "Information Gain==>", inf)
        return inf,thresh_best

    def pipeline2readyData(self, x, y):
        x1 = x.copy()
        y1 = y.copy()
        for i in range(x1.shape[1]):
            if len(np.unique(x1[:, i])) >= 3:
                _, split = self.best_split(x1[:, i], y1, i)
                self.thresholdValues[i] = split
                for j in range(len(x1[:, i])):
                    if x1[:, i][j] >= split:
                        x1[:, i][j] = 1
                    else:
                        x1[:, i][j] = 0
            else:
                continue
        return x1, y1

    def fit(self, x, y,index=0,dic={}):
        if index == 0:
            self.treeNode = [-1,]*self.treeSize
        if self.entropy(y) <= self.entropyThreshold:
            freq = maximumFreq(y)
            if freq == -1:
                self.treeNode[index] = 'C'+str(np.random.randint(0,self.classes))
            else:
                self.treeNode[index] = 'C'+str(freq)
            self.root.append(dic)
            return
        if len(dic) >= self.maxDepth:
            self.treeNode[index] = 'C'+str(maximumFreq(y))
            self.root.append(dic)
            return
        inf = -100
        split = -1
        for i in range(x.shape[1]):
            if i not in dic:
                temp = self.information_gain(self.entropy(y),x[:,i] ,y)
                if temp > inf:
                    inf = temp
                    split = i
        if split == -1:
            self.treeNode[index] = 'C'+str(maximumFreq(y))
            self.root.append(dic)
            return
        dic[split] = index2column[split]
        self.treeNode[index] = split
        x_ = x.copy()
        y_ = y.copy()
        x_left = x_[x_[:, split] == 0]
        x_right = x_[x_[:, split] == 1]
        y_left = y_[x_[:, split] == 0]
        y_right = y_[x_[:, split] == 1]
        dic_ = dic.copy()
        self.fit(x_left, y_left,2*index+1,dic_ )
        self.fit(x_right, y_right,2*index+2,dic_)

    def predict_one(self, x):
      i = 0
      while type(self.treeNode[i]) != str:
        if x[self.treeNode[i]] == 0:
          i = 2*i + 1
        elif x[self.treeNode[i]] == 1:
          i = 2*i + 2
      return float(self.treeNode[i][1:])

    def apply_thresh(self,x_):
        length = len(x_)
        x = x_.copy()
        for i in self.thresholdValues:
            divide = self.thresholdValues[i]
            for j in range(length):
                if x[:,i][j] >= divide:
                    x[:,i][j] = 1
                else:
                    x[:,i][j] = 0
        return x
        
    def predict(self, x_):
      y__ = []
      x = self.apply_thresh(x_)
      for row in x:
        y__.append(self.predict_one(row))
      return np.array(y__)

    def all_paths(self):
        li = []
        for i in tree.root:
            if i not in li:
                li.append(i)
        return li
        
```

```python
def accuracy(y_test, y_pred):
    counter = 0
    for i in range(len(y_test)):
        if y_test[i] == y_pred[i]:
            counter+=1
    return counter/len(y_test)
```

```python
def classwise_accuracy(y_test, y_pred):
    dic= {}
    for i in y_test:
        if i not in dic:
            dic.update({i: 'Class '+str(i)})
    accuracy = {}
    for i in range(len(y_test)):
        if y_test[i]!=y_pred[i]:
            if y_test[i] not in accuracy:
                accuracy.update({y_test[i]: [0,1]})
            else:
                accuracy[y_test[i]][1]+=1
        else:
            if y_test[i] not in accuracy:
                accuracy.update({y_test[i]:[1,0]})
            else:
                accuracy[y_test[i]][0]+=1
    for i in accuracy:
        accuracy[i] = accuracy[i][0] / (accuracy[i][0] + accuracy[i][1])
        print(dic[i],'==> Accuracy:',accuracy[i])
            
```

```python
def confusion_matrix(y_test, y_pred):
    import numpy as np
    counter = 0
    dict = {}
    for i in y_test:
        if i not in dict:
            dict.update({i:counter})
            counter+=1
    n = len(dict)
    arr = np.zeros((n,n))
    for i in range(len(y_test)):
        arr[dict[y_test[i]]][dict[y_pred[i]]]+=1
    return arr
```
